Plot count plots for categorical columns

```python
sns.set(style="whitegrid")
plt.figure(figsize=(14, 10))
campaign_name_order = df['Campaign Name'].value_counts().index
ax = sns.countplot(y=df['Campaign Name'], order=campaign_name_order)
plt.title('Distribution of Campaign Names', fontsize=16)
plt.xlabel('Frequency', fontsize=14)
plt.ylabel('Campaign Name', fontsize=14)
ax.set_yticklabels(ax.get_yticklabels(), rotation=0, fontsize=13)
plt.show()
```

INSIGHT: In the above visualization it tells the which campaign has the more number Reaches which will reflect the frequency of leads and the campaign which has more engagement and reach

Recommendation: Folowing the factors which have Influenced the Campaign with more leads can be applied to others so that we can boost the Enagagement

```python
#Time series analysis Leads generated overTime
plt.figure(figsize=(14, 7))
df['Created_Date'].value_counts().sort_index().plot(kind='line')
plt.title('Number of Leads Created Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Leads')
plt.show()
```

INSIGHT: From late 2023 to early 2024, there is a general upward trend in the number of leads created, indicating a growth phase. This might be due to successful marketing campaigns, increased brand awareness, or market expansion After April 2024, there is a sharp decline in the number of leads created, followed by a period of stabilization at a lower level. This decline could be due to the end of a campaign, seasonal market changes, or other external factors The highest peaks occur around February to April 2024, with the number of leads exceeding 500 during these times These periods might correspond to specific events, promotions, or market conditions that drive higher lead generation

Recommendation: Plan future marketing campaigns and lead generation efforts around the identified peak periods nvestigate the reasons for the sharp declines post-April and develop strategies to mitigate such drops in the future Understanding the factors contributing to these peaks can help replicate success in future campaigns

Bivariate Analysis

```python
#Visualize the relationship between two numeric variables
sns.barplot(x='graduation_percentage', y='work_experience', data=df)
plt.show()
```

INSIGHT: There is a noticeable trend where individuals with higher graduation percentages tend to have more work experience. Specifically, those with graduation percentages in the range of 95-100% have significantly more work experience compared to those with lower graduation percentages For graduation percentages between 50% and 85%, the average work experience remains relatively constant, hovering around 2-3 years. This suggests that within this range, graduation percentage does not have a strong influence on the amount of work experienceBeyond 85%: Beyond the 85% graduation percentage mark, there is a significant increase in work experience. This indicates that individuals who perform exceptionally well academically also tend to have more extensive work experience

Recommendation: Given the positive correlation between high academic performance and extensive work experience, consider targeting marketing efforts towards individuals with higher graduation percentages (85% and above) making sure that they are alligned with their Ambition For leads with average graduation percentages, emphasize the value of practical experience and how your offerings can enhance their career prospects. Showcase success stories of individuals who have leveraged their work experience to achieve career growth

```python
# Group by Improvements and count the number of leads
improvements_count = df.groupby('Improvements').size().reset_index(name='Lead Count')
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='Improvements', y='Lead Count', data=improvements_count, palette='viridis')
plt.title('Count of Leads by Improvements')
plt.xlabel('Improvements')
plt.ylabel('Lead Count')
plt.show()
```

INSIGHT:significantly higher number of leads are in the category where no improvements are needed (0) compared to the number of leads that do require improvements(1)
Count of Leads Without Improvements: Approximately 28,000 leads
Count of Leads With Improvements: Approximately 4,000 leads

Recommendation:Develop targeted improvement strategies for the leads that do require improvements. This could involve additional training, follow-ups, or providing more detailed information to these leads to bring them up

```python
# 2. Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=df['platform'], y=df['graduation_percentage'])
plt.title('Bar Plot of Graduation Percentage by Platform')
plt.xlabel('Platform')
plt.ylabel('Average Graduation Percentage')
plt.xticks(rotation=45)
plt.show()
```

Recommendation:Given that LinkedIn leads have the highest median graduation percentage, it might be beneficial to focus more marketing efforts on LinkedIn for campaigns targeting individuals with higher academic achievements
